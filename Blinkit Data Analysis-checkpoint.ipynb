# Blinkit Data Analysis

```python
#Importing requried librabries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#loading dataset
data = pd.read_csv("blinkit_data.csv")
```

```python
#Data cleaning
data['Item Fat Content'] = data['Item Fat Content'].replace({'LF' : 'Low Fat',
                                                             'low fat' : 'Low Fat',
                                                             'reg' : 'Regular'})
```

```python
#Adding Age column for business requirement
data['Outlet Age'] = 2025 - data['Outlet Establishment Year']
```

# 3.Chart Requirements

### 1.Compare total sales by item type

```python
# Calculating total sales by Item Type
sales_by_type = data.groupby('Item Type')['Sales'].sum().reset_index()

# Sortiing by sales descending
sales_by_type = sales_by_type.sort_values(by='Sales', ascending=False)

# Ploting the bar chart
plt.figure(figsize=(12, 6))
sns.barplot(data=sales_by_type, x='Item Type', y='Sales', palette='mako')
plt.xticks(rotation=90)
plt.title('Total Sales by Item Type (High to Low)')
plt.xlabel('Item Type')
plt.ylabel('Total Sales')
plt.tight_layout()
plt.show()
```

### Total Sales by Item Type (High to Low):

#### • Top Sellers: Fruits and Vegetables and Snack Foods are the highest-grossing categories, indicating consistent customer demand for these essential and impulse items.

#### • Household & Frozen Foods: These rank just below the top, showing that non-edible daily-use items and convenience foods also significantly contribute to total revenue.

#### • Low Performing Items: Categories like Breakfast, Seafood, and Starchy Foods generate the least sales, suggesting either limited demand or inventory presence.

#### • Balanced Contribution: While food items dominate the top, the mix of categories shows that both edible and non-edible product types drive sales performance.

### 2.Total Sales by Fat Content

```python
sales_by_fat = data.groupby('Item Fat Content')['Sales'].sum()

# plotting the pie chart

plt.pie(sales_by_fat, labels=sales_by_fat.index, autopct='%1.1f%%', startangle=100, colors=plt.cm.Paired.colors)
plt.title('Total Sales Distribution by Item Fat Content')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.tight_layout()
plt.show()
```

#### Total Sales By Fat Content:

#### • Low Fat Dominance: Products labeled Low Fat contribute 64.6% of total sales, indicating a strong consumer preference for healthier or diet-conscious options.

#### • Regular Fat Share: Regular items account for 35.4%, which still represents a substantial portion, suggesting mixed demand across health-focused and traditional products.

#### • Balanced Inventory Needed: While Low Fat leads, maintaining a range of Regular fat products remains important to serve a broader customer base.

#### • Potential for Health Marketing: The strong sales in Low Fat items highlight an opportunity to further promote health-oriented product lines.

### 3.Fat Content by Outlet for Total Sales

```python
fat_outlet = data.groupby(['Outlet Identifier', 'Item Fat Content'])['Sales'].sum().reset_index()

plt.figure(figsize=(14, 6))
sns.barplot(data=fat_outlet, x='Outlet Identifier', y='Sales', hue='Item Fat Content', palette='Set2')
plt.title('Total Sales by Fat Content Across Outlets')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

### Total Sales by Fat Content Across Outlets:

#### • Low Fat Dominance: In every outlet, Low Fat items consistently generate higher total sales than Regular fat items.

#### • Top Performing Outlets: OUT013, OUT018, OUT035, and OUT045 show the highest total sales for both fat types, indicating strong outlet performance.

#### • Sales Gap: The difference between Low Fat and Regular item sales is most pronounced in OUT010 and OUT019, suggesting specific customer preferences.

#### • Uniform Trends: The preference for Low Fat items is stable across nearly all outlets, pointing to a company-wide trend in consumer demand.


### 4.Track sales by outlet location type 

```python
loc_sales = data.groupby('Outlet Location Type')['Sales'].sum().reset_index().sort_values(by='Sales', ascending=False)

plt.figure(figsize=(8, 5))
sns.barplot(data=loc_sales, x='Outlet Location Type', y='Sales',hue='Outlet Location Type', palette='rocket')
plt.title('Total Sales by Outlet Location Type')
plt.tight_layout()
plt.show()
```

### Total Sales by Outlet Location Type

### 5.Compare sales by outlet size & type 

```python
size_type = data.pivot_table(index='Outlet Size', columns='Outlet Type', values='Sales', aggfunc='sum')

plt.figure(figsize=(8, 6))
sns.heatmap(size_type, annot=True, fmt=".0f", cmap='YlGnBu')
plt.title('Sales by Outlet Size and Type')
plt.tight_layout()
plt.show()
```

### Sales by Outlet Size & Type

#### • Small Supermarket Type 1 outlets have the highest sales (362,603), indicating strong performance.

#### • Medium-sized Supermarket Type 2 and 3 show decent sales (~130K each), while Grocery Stores perform lowest across all sizes.

#### • High outlet size contributes more in Supermarket Type 1 but not in other types.

#### • Supermarket Type 1 dominates sales across all sizes, making it the most profitable outlet type.

#### • Larger outlet size does not necessarily lead to higher sales—small-sized outlets outperform medium and high in key categories.

### 6.Identify high-revenue product categories

### Top Performers:

#### • Household and Frozen Foods also perform well, making them key revenue drivers.

### Mid-Level Categories:

#### • These are consistent earners with potential for growth.

### Low-Revenue Categories:

#### • Seafood, Breakfast, and Starchy Foods are the lowest revenue contributors.

#### • These may need strategic promotion or inventory review.
